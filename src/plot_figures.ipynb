# Figure 5

```python
from utils.cal_metrics import *
import pandas as pd
import numpy as np
from tqdm import tqdm
import torch
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scanpy as sc
from sklearn.preprocessing import scale
import scipy
lincs_adata = sc.read('data/lincs_adata.h5ad')
sc.pp.normalize_total(lincs_adata)
sc.pp.log1p(lincs_adata)
drug_df = lincs_adata.obs.condition.value_counts().to_frame().iloc[1:101]
drug_df['split'] = ""
for drugi in range(100):
    if lincs_adata.obs.drug_split_0[lincs_adata.obs.condition==drug_df.index[drugi]].iloc[0]=='test':
        split_idx='drug_split_0'
    elif lincs_adata.obs.drug_split_1[lincs_adata.obs.condition==drug_df.index[drugi]].iloc[0]=='test':
        split_idx='drug_split_1'
    elif lincs_adata.obs.drug_split_2[lincs_adata.obs.condition==drug_df.index[drugi]].iloc[0]=='test':
        split_idx='drug_split_2'
    elif lincs_adata.obs.drug_split_3[lincs_adata.obs.condition==drug_df.index[drugi]].iloc[0]=='test':
        split_idx='drug_split_3'
    elif lincs_adata.obs.drug_split_4[lincs_adata.obs.condition==drug_df.index[drugi]].iloc[0]=='test':
        split_idx='drug_split_4'
    drug_df.iloc[drugi,-1] = split_idx
drug_df['deg_cor'] = 0
drug_df['matrix_cor'] = 0
drug_df['cell_cnt'] = 0
```

```python
drug_name = 'lapatinib'
split_key = drug_df.loc[drug_name,'split']
folder='data/result/'+ split_key + '/Cross_0_batch5000'
common_folder = 'data/result/'+ split_key + '/common'
total_y_pred0 = torch.load(folder+'/total_y_pred.pkl')
total_y_true0 = torch.load(common_folder+'/total_y_true.pkl')
total_x0 = torch.load(common_folder+'/total_x.pkl')
test_adata = lincs_adata[(lincs_adata.obs[split_key] == 'test') & (lincs_adata.obs['dose']!= 0.0)]

subset_idx = list(test_adata.obs.condition==drug_name)
obs = test_adata.obs.copy()
obs['condition_dose'] = obs.cov_drug_dose_name.apply(lambda x: '_'.join(str(x).split('_')[1:]))
obs_subset = obs[subset_idx]
total_y_pred = total_y_pred0[subset_idx]
total_y_true = total_y_true0[subset_idx]
total_x = total_x0[subset_idx]

# obtain the exprerssion table split by cell_type in one drug
obs_key='cell_type'
res_dict = {}
column_name = obs_subset[obs_key]  
for index, key in tqdm(enumerate(column_name)):
    if key not in res_dict:
        res_dict[key] = {'x': [], 'y_true': [], 'y_pred': []}
    res_dict[key]['x'].append(total_x[index])
    res_dict[key]['y_true'].append(total_y_true[index])
    res_dict[key]['y_pred'].append(total_y_pred[index])
correlation_coefficients = []
FC_true, FC_pred = {}, {}
for key, values in tqdm(res_dict.items()):
    x_list = np.array(values['x']).mean(axis=0)
    y_true_list = np.array(values['y_true']).mean(axis=0)
    y_pred_list = np.array(values['y_pred']).mean(axis=0)
    y_true_over_x = y_true_list - x_list
    y_pred_over_x = y_pred_list - x_list
    FC_true[key] = y_true_over_x
    FC_pred[key] = y_pred_over_x
pd_true = pd.DataFrame(FC_true.values(), index=FC_true.keys(), columns=test_adata.var_names)
pd_pred = pd.DataFrame(FC_pred.values(), index=FC_pred.keys(), columns=test_adata.var_names)
# pd_true = pd.DataFrame(scale(pd_true, axis=0), columns=pd_true.columns, index=FC_true.keys())
# pd_pred = pd.DataFrame(scale(pd_pred, axis=0), columns=pd_pred.columns, index=FC_pred.keys())
# calculate logfc PCC of each gene
correlation_matrix = pd_true.corrwith(pd_pred)
correlation_df = pd.DataFrame(correlation_matrix)
correlation_df.columns = ['Correlation_Coefficient']
correlation_df = correlation_df.sort_values(by='Correlation_Coefficient', ascending=False)

# calculate deg and select top 25 up and down degs to see the cor
lincs_adata_subset = lincs_adata[(lincs_adata.obs.condition==drug_name) | (lincs_adata.obs.condition=='DMSO')]
lincs_adata_subset.uns['rank_genes_groups'] = {}
sc.tl.rank_genes_groups(
    lincs_adata_subset,
    groupby='condition', 
    reference='DMSO',
    rankby_abs=True
)
df = sc.get.rank_genes_groups_df(lincs_adata_subset,group=drug_name)
df.set_index('names',inplace=True)
df['cor'] = correlation_df.Correlation_Coefficient[df.index]
topgenes = list(df[(df.scores<0)].index[:25]) + list(df[(df.scores>0)].index[:25])
deg_cor_mean = df.loc[topgenes,'cor'].mean()
print(f'The mean deg logFC PCC of drug {drug_name} is {deg_cor_mean}')
print(f'The deg logFC PCC of drug by the whole matrix {drug_name} is {np.corrcoef(pd_true.loc[:, topgenes].values.flatten(), pd_pred.loc[:, topgenes].values.flatten())[0,1]}')
```

```python
all_values = np.concatenate((pd_true.loc[:, topgenes].values.flatten(), pd_pred.loc[:, topgenes].values.flatten()))
vmin_value = all_values.min()
vmax_value = all_values.max()
fig, axes = plt.subplots(1, 2, figsize=(24, 6))
g1 = sns.heatmap(pd_true.loc[:, topgenes], vmin=vmin_value, vmax=vmax_value, cmap='RdBu_r', center=0, annot=False, fmt='.2f', ax=axes[0], cbar=False)
axes[0].set_title('Groud Truth',fontsize=20)
g2 = sns.heatmap(pd_pred.loc[:, topgenes], vmin=vmin_value, vmax=vmax_value, cmap='RdBu_r', center=0, annot=False, fmt='.2f', ax=axes[1], cbar=False)
axes[1].set_title('CrossDiT Prediction',fontsize=20)
cbar = fig.colorbar(g1.collections[0], ax=axes.ravel().tolist(), shrink=0.6, aspect=20)
cbar.outline.set_visible(False)
# plt.subplots_adjust(wspace=0.3)
# plt.savefig(f'{drug_name}_heatmaps.png')
plt.savefig('data/Figures/Figure5.pdf', format='pdf', bbox_inches='tight')
```
